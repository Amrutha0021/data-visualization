## 1.5 Vectorized arithmetic & broadcasting

```
prices = np.array([100, 250, 80, 400])
qty = np.array([2, 1, 5, 1])

print("total per item:", prices * qty)
print("grand total   :", (prices * qty).sum())

# Broadcasting: apply a flat 10% discount to every price
discounted = prices * 0.9
print("discounted    :", discounted)

# Broadcasting a row vector across a matrix
matrix = np.ones((3, 4))
row = np.array([1, 2, 3, 4])
print("broadcast add:\n", matrix + row)
```

## 1.6 Ufuncs & aggregations

grid (weeks x days), then find the total steps per week and which single day had the most steps overall.

```
np.random.seed(42)
daily_steps = np.random.randint(3000, 12000, size=28)

# TODO: reshape into (4 weeks, 7 days)
weekly_grid = None

# TODO: total steps per week (axis that sums across days)
steps_per_week = None

# TODO: index (0-27) of the single best day, and its step count
best_day_index = None
best_day_steps = None

print(weekly_grid)
print("Per-week totals:", steps_per_week)
print("Best day:", best_day_index, "with", best_day_steps, "steps")
```

## 🎯 Mini-challenge — Pandas capstone
Using `employees_clean`, in a single chain (or a few steps):
1. Fill missing `salary` with the department's mean salary (not the overall mean).
2. Group by `department` and compute the average `salary` and headcount.
3. Sort the result by average salary, descending.

```
# TODO: fill missing salary with each row's DEPARTMENT mean, then group + sort
df = employees_clean.copy()

df["salary"] = None  # hint: groupby("department")["salary"].transform("mean")

summary = None  # groupby department -> avg_salary, headcount
summary_sorted = None  # sort by avg_salary descending

summary_sorted
```

### ✅ Solution

```
df = employees_clean.copy()
df["salary"] = df["salary"].fillna(df.groupby("department")["salary"].transform("mean"))

summary = df.groupby("department").agg(
    avg_salary=("salary", "mean"),
    headcount=("name", "size"),
)
summary_sorted = summary.sort_values("avg_salary", ascending=False)

summary_sorted
```
